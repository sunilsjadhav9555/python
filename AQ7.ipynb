```python
7.Write a Python program that keeps asking the user for a positive integer. If the user
enters a negative number or zero, it should continue asking until a valid number is
provided.
```

```python
while True:
    num = int(input("Enter a positive integer: "))
    if num > 0:
        print(f"Thank you! You entered {num}.")
        break  
    else:
        print("Invalid input. Please enter a positive integer.")
        continue 
```
