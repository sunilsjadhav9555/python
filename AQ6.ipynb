```python
6.Write a Python code that processes a list of strings and ignores any empty strings
or strings with only whitespace using the continue statement.
a. ["hello", " ", "", "world", " ", "python"]
b. Output= Hello, World, python
```

```python
a = ["hello", " ", "", "world", " ", "python"]

filtered_words = []

for word in a:
    
    if word.strip() == "":
        continue  
        
    filtered_words.append(word.strip())

output = ", ".join(filtered_words)
print(output)
```
